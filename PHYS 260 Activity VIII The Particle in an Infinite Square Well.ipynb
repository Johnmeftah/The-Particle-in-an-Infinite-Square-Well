```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for the infinite potential well
n = 5  # Quantum number
L = 1  # Width of the well, arbitrary units

# Define the position space from 0 to L
x = np.linspace(0, L, 1000)

# Defining the wavefunction for n=5
psi_n = np.sqrt(2/L) * np.sin(n * np.pi * x / L)

# Calculating the probability density function
probability_density = psi_n**2

# Enhancing the plot with additional styling
# Defining the figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Ploting the probability density with a thicker line and a different color
ax.plot(x, probability_density, label=f'n={n}', linewidth=2, color='royalblue')

# Adding a vertical line in the middle of the plot at x = L/2
ax.axvline(L/2, color='black', linestyle='--', linewidth=1)

# Adding title and labels with a larger font
ax.set_title('Probability Density for a Particle in a 1D Infinite Potential Well (n=5)', fontsize=15)
ax.set_xlabel('Position x/L', fontsize=12)
ax.set_ylabel('Probability Density', fontsize=12)

# Setting larger tick labels for better readability
ax.tick_params(axis='both', which='major', labelsize=10)

# Adding a legend with a shadow
ax.legend(loc='upper right', fontsize=12, shadow=True)

# Adding grid lines with a lighter color and dashed style
ax.grid(True, linestyle='--', color='grey', alpha=0.7)

# Adding a fancy style to the plot (because why not?)
plt.style.use('ggplot')

# Save the plot as a high-resolution PNG file
plt.savefig("activity_VIII.png", dpi=500)
plt.show()
```
